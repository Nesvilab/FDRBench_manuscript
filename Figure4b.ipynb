```r
library(tidyverse)
library(ggpubr)
get_ratio=function(fdp_file, dataset_name="",fdr_decimal_place=1){
    a <- read_csv(fdp_file)
    cat("The number of discoveries passed 1% FDR:", a %>% filter(q_value<=0.01) %>% nrow,"\n")
    cat("The number of discoveries passed 1% FDP (paired method):", a %>% filter(FDP_1B<=0.01) %>% nrow,"\n")
    cat("Improvement:",(a %>% filter(q_value<=0.01) %>% nrow)/(a %>% filter(FDP_1B<=0.01) %>% nrow) - 1,"\n")

    max_score_FDP_cutoff <- max(a$score[a$FDP_1B<=0.01])
    cat("The maximum score of discoveries passed 1% FDP:", max_score_FDP_cutoff,"\n")
    cat("The number of discoveries passed 1% FDR:", a %>% filter(q_value<=0.01) %>% nrow,"\n")
    cat("The number of discoveries passed 1% FDP (paired method):", a %>% filter(score<=max_score_FDP_cutoff) %>% nrow,"\n")
    n_original_discoveries <- a %>% filter(q_value<=0.01) %>% nrow
    n_fdp_discoveries <- a %>% filter(score<=max_score_FDP_cutoff) %>% nrow
    ratio <- (n_original_discoveries/n_fdp_discoveries) - 1
    cat("Improvement:",(a %>% filter(q_value<=0.01) %>% nrow)/(a %>% filter(score<=max_score_FDP_cutoff) %>% nrow) - 1,"\n")
    if(fdr_decimal_place==1){
        fdp_qvalue001 <- a %>% filter(q_value<=0.01) %>% arrange(desc(q_value),desc(score)) %>% filter(row_number()==1) %>% mutate(ratio=sprintf("%.1f%%",FDP_1B*100))
    }else if(fdr_decimal_place==2){
        fdp_qvalue001 <- a %>% filter(q_value<=0.01) %>% arrange(desc(q_value),desc(score)) %>% filter(row_number()==1) %>% mutate(ratio=sprintf("%.2f%%",FDP_1B*100))
    }else{
        fdp_qvalue001 <- a %>% filter(q_value<=0.01) %>% arrange(desc(q_value),desc(score)) %>% filter(row_number()==1) %>% mutate(ratio=sprintf("%.4f%%",FDP_1B*100))
    }

    aa_tmp <- a %>% filter(score<=max_score_FDP_cutoff,FDP_1B<=0.01) 
    x <- data.frame(dataset_name=dataset_name,
                    oiginal_discoveries=n_original_discoveries,
                    fdp_discoveries=n_fdp_discoveries,
                    ratio=ratio, 
                    fdp=fdp_qvalue001$ratio, 
                    fdp_cutoff=max(aa_tmp$FDP_1B),
                    stringsAsFactors = FALSE)
    return(x)
}

## DIA-NN: precursor level
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD042704_Human_Astral/peptide_level_fasta/results/nf-teirex-dia/diann/peptide-diann_fdp_precursor.csv"
g1 <- get_ratio(fdp_file,dataset_name="human-astral")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD028735_LFQ_Orbitrap_AIF_Human/peptide_level_fasta/results/nf-teirex-dia/diann/peptide-diann_fdp_precursor.csv"
g2 <- get_ratio(fdp_file,dataset_name="human-qe")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD028735_LFQ_TTOF5600_Human/peptide_level_fasta/results/nf-teirex-dia/diann/peptide-diann_fdp_precursor.csv"
g3 <- get_ratio(fdp_file,dataset_name="human-tripletof")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/MSV000084000/peptide_level_fasta/results/nf-teirex-dia/diann/peptide-diann_fdp_precursor.csv"
g4 <- get_ratio(fdp_file,dataset_name="yeast-lumos")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD012988_Mouse/peptide_level_fasta/results/nf-teirex-dia/diann/peptide-diann_fdp_precursor.csv"
g5 <- get_ratio(fdp_file,dataset_name="mouse-qe")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD017703_timsTOF_Human/peptide_level_fasta/diann/peptide-diann_fdp_precursor.csv"
g6 <- get_ratio(fdp_file,dataset_name="human-timstof2")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD041421_timsTOF_Human/peptide_level_fasta/diann/peptide-diann_fdp_precursor.csv"
g7 <- get_ratio(fdp_file,dataset_name="human-timstof1")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD023325_100Cell/peptide_level_fasta/results/nf-teirex-dia/diann/peptide-diann_fdp_precursor.csv"
g8 <- get_ratio(fdp_file,dataset_name="100cell-eclipse")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD023325_1Cell/peptide_level_fasta/results/nf-teirex-dia/diann/peptide-diann_fdp_precursor.csv"
g9 <- get_ratio(fdp_file,dataset_name="1cell-eclipse")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD034525_Human_AD/peptide_level_fasta/results/nf-teirex-dia/diann/peptide-diann_fdp_precursor.csv"
g10 <- get_ratio(fdp_file,dataset_name="human-lumos")

dat_pep <- bind_rows(g1,g2,g3,g4,g5,g6,g7,g8,g9,g10) %>% mutate(level="precursor")


## DIA-NN: protein level
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD042704_Human_Astral/protein_level/results/nf-teirex-dia/diann/protein-diann_fdp_protein.csv"
g1 <- get_ratio(fdp_file,dataset_name="human-astral")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD028735_LFQ_Orbitrap_AIF_Human/protein_level/results/nf-teirex-dia/diann/protein-diann_fdp_protein.csv"
g2 <- get_ratio(fdp_file,dataset_name="human-qe")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD028735_LFQ_TTOF5600_Human/protein_level/results/nf-teirex-dia/diann/protein-diann_fdp_protein.csv"
g3 <- get_ratio(fdp_file,dataset_name="human-tripletof")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/MSV000084000/protein_level/results/nf-teirex-dia/diann/protein-diann_fdp_protein.csv"
g4 <- get_ratio(fdp_file,dataset_name="yeast-lumos")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD012988_Mouse/protein_level/results/nf-teirex-dia/diann/protein-diann_fdp_protein.csv"
g5 <- get_ratio(fdp_file,dataset_name="mouse-qe")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD017703_timsTOF_Human/protein_level/diann/protein-diann_fdp_protein.csv"
g6 <- get_ratio(fdp_file,dataset_name="human-timstof2")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD041421_timsTOF_Human/protein_level/diann/protein-diann_fdp_protein.csv"
g7 <- get_ratio(fdp_file,dataset_name="human-timstof1")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD023325_100Cell/protein_level/results/nf-teirex-dia/diann/protein-diann_fdp_protein.csv"
g8 <- get_ratio(fdp_file,dataset_name="100cell-eclipse")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD023325_1Cell/protein_level/results/nf-teirex-dia/diann/protein-diann_fdp_protein.csv"
g9 <- get_ratio(fdp_file,dataset_name="1cell-eclipse")
fdp_file <- "/data/scratch/bwen1/project/2023_entrapment_fdr/fdr_evaluation/diann/PXD034525_Human_AD/protein_level/results/nf-teirex-dia/diann/protein-diann_fdp_protein.csv"
g10 <- get_ratio(fdp_file,dataset_name="human-lumos")


dat_pro <- bind_rows(g1,g2,g3,g4,g5,g6,g7,g8,g9,g10) %>% mutate(level="protein")

dat <- bind_rows(dat_pep,dat_pro)
dat$dataset_name <- factor(dat$dataset_name, 
                           levels=c("human-astral","human-qe","human-tripletof","yeast-lumos","mouse-qe","human-timstof2","human-timstof1","100cell-eclipse","1cell-eclipse","human-lumos"))
dat %>% write_tsv("diann_fdp_vs_fdr.tsv")


options(jupyter.plot_mimetypes = "image/png")
options(repr.plot.width = 3.5*3, repr.plot.height = 3.5*1.3, repr.plot.res = 200)
gg <- ggplot(dat,aes(x=dataset_name,y=ratio,fill=dataset_name)) + 
    geom_bar(stat="identity",width=0.6) + 
    ## show the ratio on the top of the bar
    geom_text(aes(label=fdp,y=1.2*max(dat$ratio)), vjust=-0, size=3) +
    geom_text(aes(label=paste(sprintf("%.1f%%",100*ratio),"\n",oiginal_discoveries,"\n",fdp_discoveries,sep="")), vjust=-0.1, size=3,lineheight=0.9) +
    theme_bw()+
    #geom_segment(x=0.01,xend=0.01,y=0,yend=0.01,linewidth=0.3,color="blue",linetype=2)+
    theme_pubr(base_size = 12,border = TRUE)+
    ylab("Inflation rate") + 
    xlab("Dataset") +
    ylim(0,1.2*max(dat$ratio))+
    scale_y_continuous(labels = scales::percent)+
    theme(legend.position = "none",axis.text.x = element_text(angle = 60, hjust = 1),plot.margin = unit(c(0.1,0.3,0.2,0.1), "cm"),
          legend.text=element_text(size=12),
          legend.title=element_text(size=12),
          strip.text = element_text(size=12))+#,strip.background = element_rect(size=0.5))+
    facet_wrap(~level)

print(gg)

pdf("Figure-diann_fdp-fdr-comparison.pdf",width = 3.5*3, height = 3.5*1.3)
print(gg)
dev.off()
```
